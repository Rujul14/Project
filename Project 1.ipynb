```python
import matplotlib.pyplot as plt
import pandas as pd 
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
dataset = pd.read_csv(r"C:\Users\Rujul\Desktop\ML Project\time_series_2019-ncov-Confirmed.csv")
```

```python
pd.set_option('display.max_rows',None)
dataset
```

```python
df = dataset.drop(['Province/State','Lat','Long'],axis=1)
df.describe()
```

```python
keys=df.keys()
keys=keys[1:]
keys
```

```python
#To drop columns where 70% number of confirmed cases is zero on a particular date
for i in keys:
    count=0
    for row in range(487):
        if(df[i][row]==0):
            count+=1
    per=count/487
    if(per>0.71):
        df = df.drop([i],axis=1)
```

```python
df
```

```python
#Task 1 - to remove duplicate countries and get disctinct country values. .sum is used to sum up the values for similar countries
df1=df.groupby('Country/Region').sum()
df1 = df1.reset_index()
```

```python
df1
```

```python
#Task 2 - Plot country wise covid cases
plt.figure(figsize=(100,50))
ax=plt.subplot(1,1,1)
x=range(171)
y=df1.iloc[:,-1]
ax.set_xticks(range(171))
# ax.set_yticks(y)
ax.set_xticklabels(df1.iloc[:,0], rotation=60,fontsize=20)
plt.yticks(fontsize=40)
plt.plot(x,y,c='purple',lw=5)
plt.title('Country wise cases',fontsize=70)
plt.xlabel('Countries',fontsize=70)
plt.ylabel('Total COVID Cases',fontsize=60)
plt.grid()
plt.show()
```

```python
x_val=df1.keys()
x_val = x_val[1:]
x_val
```

```python
y_val=[]
for i in x_val:
    y_val.append(df[i].sum())
print(y_val)
```

```python
#Task 2 - Plot date wise covid cases
plt.figure(figsize=(60,30))
ax=plt.subplot(1,1,1)
x=range(len(x_val))
y=y_val
# ax.set_yticks(y_val)
plt.yticks(fontsize=20)
ax.set_xticks(x)
ax.set_xticklabels(x_val, rotation=60, fontsize=30)
plt.plot(x,y, marker='o',c='purple', markersize=30)
plt.title('Datewise wise cases',fontsize=70)
plt.xlabel('Dates',fontsize=70)
plt.ylabel('Total COVID Cases',fontsize=55)
plt.grid()
plt.show()
```
